1 What is NumPy, and why is it widely used in Python

  ans - NumPy is a core Python library used for numerical computing. It provides a powerful data structure called the N-dimensional array (ndarray) and a collection of fast mathematical functions that operate on these arrays.

It also includes tools for:

Linear algebra (matrices, decompositions, etc.)

2  How does broadcasting work in NumPy

  ans - Broadcasting in NumPy is a set of rules that lets it perform arithmetic operations on arrays of different shapes without explicitly copying or reshaping data.

It basically means: NumPy tries to “stretch” smaller arrays so they behave like larger ones during an operation—as long as the shapes are compatible.
Core idea of broadcasting

When you do something like:

a + b

NumPy compares the shapes of a and b from right to left and applies rules to see if they can align.

Instead of physically duplicating data, it virtually repeats values when needed.

21 Why is NumPy’s array slicing faster than Python’s list slicing

  ans - NumPy array slicing in NumPy is faster than Python list slicing mainly because of how data is stored and how slicing is implemented at the memory level.

1. Contiguous memory layout

NumPy arrays store data in a continuous block of memory, unlike Python lists which store references (pointers) to objects scattered in memory.

So when slicing:

NumPy just adjusts a view of memory
Python lists must handle object references

👉 Less overhead = faster slicing
2. Slicing returns a view, not a copy (usually)

In NumPy:

```python
a = np.array([1, 2, 3, 4])
b = a[1:3]
```

11 Perform matrix multiplication using NumPy

```python
import numpy as np

# Define two matrices
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

# Matrix multiplication
result = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result:\n", result)
```
